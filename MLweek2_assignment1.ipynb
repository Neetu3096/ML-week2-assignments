```python
# Import libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix
)
```

PART A: DATASET UNDERSTANDING

```python
# Q1. Load the dataset and display the first five records.
data = pd.read_csv("Dataset 2.csv")

print("First Five Records:")
print(data.head())
```

```python
#Q2. Determine the number of rows and columns in the dataset.
print("\nRows and Columns:")
print(data.shape)
```

```python
#Q3. Display all column names.
print("\nColumn Names:")
print(data.columns)
```

```python
#Q4. Identify numerical and categorical features.
print("\nNumerical Columns:")
print(
data.select_dtypes(
include=np.number
).columns
)

print("\nCategorical Columns:")
print(
data.select_dtypes(
exclude=np.number
).columns
)
```

```python
#Q5. Check whether the dataset contains missing values.
print("\nMissing Values:")
print(data.isnull().sum())
```

Part B: Exploratory Data Analysis

```python
#Q6. Calculate the average age of users.
print("\nAverage Age:")
print(data["Age"].mean())
```

```python
#Q7. Determine the average watch hours per week.
print("\nAverage Watch Hours:")
print(data["WatchHoursPerWeek"].mean())
```

```python
#Q8. Find the average monthly spending of users.
print("\nAverage Monthly Spend:")
print(data["MonthlySpend"].mean())
```

```python
#Q9. Count the number of users in each subscription category.
print("\nSubscription Count:")
print(
data["SubscriptionType"].value_counts())
```

```python
#Q10. Determine the percentage of users who renewed their subscriptions.
renewal = ((data["SubscriptionRenewed"]=="Yes").mean()*100)
print("\nRenewal Percentage:")
print(round(renewal,2))
```

Part C: Data Preparation

```python
#Q11. Convert categorical features into numerical form.
encoder = LabelEncoder()

columns = ["Gender","SubscriptionType","FavoriteGenre","SubscriptionRenewed"]

for col in columns:
    data[col] = encoder.fit_transform(data[col])

print("\nEncoded Data:")
print(data.head())
```

```python
#Q12. Define the feature set (X) and target variable (y) for subscription renewal prediction.
X = data.drop(["SubscriptionRenewed","MonthlySpend"],axis=1)
y = data["SubscriptionRenewed"]
```

```python
#Q13. Split the dataset into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)

print("\nTraining Shape:")
print(X_train.shape)

print("Testing Shape:")
print(X_test.shape)
```

Part D: Decision Tree Classification

```python
#Q14. Train a Decision Tree model to predict whether a user will renew their subscription.
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
```

```python
#Q15. Evaluate the model using accuracy.
pred_dt = dt.predict(X_test)
acc_dt = accuracy_score(y_test,pred_dt)

print("\nDecision Tree Accuracy:")
print(acc_dt)
```

```python
#Q16. Generate and interpret the confusion matrix.
cm = confusion_matrix(y_test,pred_dt)

print("\nConfusion Matrix:")
print(cm)
```

Part E: K-Nearest Neighbors (KNN)

```python
#Q17. Train a KNN classifier with K = 5.
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
pred_knn = knn.predict(X_test)
```

```python
#Q18. Compare the accuracy of KNN with the Decision Tree model.
acc_knn = accuracy_score(y_test,pred_knn)
print("\nKNN Accuracy:")
print(acc_knn)
print("\nComparison")
if acc_dt > acc_knn:
    print("Decision Tree Better")
else:
    print("KNN Better")
```

Part F: Linear Regression

```python
#Q19. Train a Linear Regression model to predict monthly spending.
X_reg = data.drop("MonthlySpend",axis=1)
y_reg = data["MonthlySpend"]
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg,y_reg,test_size=0.20,random_state=42)

lr = LinearRegression()

lr.fit(X_train_reg,y_train_reg)
```

```python
#Q20. Predict the monthly spending for a new user and interpret the result.
new_user = [[751,25,1,2,18,2,3,12,1]]

prediction = lr.predict(new_user)

print("\nPredicted Monthly Spend:")
print(prediction[0])
```
